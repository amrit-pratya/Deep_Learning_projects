```python
corpus = """Although the bridge was damaged the ambulance driver drove the vehicle across the bridge to save the patient's life."""
```

```python
import nltk
```

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
```
